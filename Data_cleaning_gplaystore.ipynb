```python
#import pandas library and load the dataset
import pandas as pd
df = pd.read_csv("gplaystore.csv")
df.head()
```

```python
#observe and explore  the data
df.info() #shows information about the data types
df.head() #show the frist 5 rows of data
```

```python
#First step is to deal with null values, there many ways like imputation and removal. usually removal if theres a large amount of data missing but, imputation if a small percantage is missing.
df.isnull().sum()
```

```python
#the chart above shows the number of missing rows in each column
df.dropna(inplace=True)
df.isnull().sum()
```

```python
#Next im going to check for Duplicated rows
df.duplicated().sum()
```

```python
#474 duplicates
df.drop_duplicates(inplace=True)
#verify
df.duplicated().sum()
```

```python
df.duplicated().sum() #double check to see any duplicated rows
```

```python
#I noticed the cells in the Install columns have a '+' at the end
df['Installs'] = df['Installs'].str.rstrip('+')

df.head()
```

```python

#Next im going to  convert the data type to the approrpiate data type.
#For installs Im going to have to strip the '+' and change that column into a numerical data type as well
df['Rating'] = pd.to_numeric(df['Rating'],errors ='coerce' )
df['Reviews'] = pd.to_numeric(df['Reviews'],errors='coerce')
df['Installs'] = pd.to_numeric(df['Installs'],errors='coerce')
df['Price'] = pd.to_numeric(df['Price'].str.replace('$', '', regex=False),errors='coerce')




```

```python
#There are apps that have different storage sizes depending on the device in this data set. So the cell will decribe it as "Varies with devices" we want to remove them since this column will be converted into a numeric.
df['Size'] = df['Size'].replace('Varies with device',pd.NA) #Every Varies with device is null now
df.dropna(inplace=True) #drop the null values
df.isnull().sum() #check for nulls

```

```python
#Check and verify to see if the data has changed
df.info()
```

```python
#for installs since i stripped the + , Ill haveto change the data type also
df['Installs'] = df['Installs'].astype(str)
df['Installs'] = df['Installs'].str.replace(',', '', regex=False)  # Remove commas
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')
#Rename the "Size" column to "Size (megabytes)" and remove the "M" from the data within that column
df.rename(columns={'Size': 'Size (megabytes)'}, inplace=True)
# Convert 'Size (megabytes)' to string before applying str.replace
df['Size (megabytes)'] = df['Size (megabytes)'].astype(str)
df['Size (megabytes)'] = df['Size (megabytes)'].str.replace('M', '', regex=False)
#Very small amount of rows had k for kilobytes so were also goin to remove the k
df['Size (megabytes)'] = df['Size (megabytes)'].str.replace('k', '', regex=False)
```

```python
#change the data type of the Size cell
df['Size (megabytes)'] = pd.to_numeric(df['Size (megabytes)'], errors='coerce')
df['Size (megabytes)'] = df['Size (megabytes)'].apply(lambda x: x / 1024 if x > 1000 else x)
```

```python
# The columns Last updated, current ver, and android ver arent relevant so Imm doing to drop them.The 'Review' column was also the exact as the Reviews 'column' column duplicate.
df.drop(['Last Updated', 'Current Ver', 'Android Ver'], axis=1, inplace=True, errors='ignore')
#Check and verify to see if the data has changed
df.info()
```

```python
df.head()
```

```python
#Lastly checck for columns that wouldnt be useful or "unnoffical" duplicates
#Category and genre are the exact same column just named differently so were going to delete Genres
df.drop('Genres', axis=1, inplace=True)
```

```python
#The "Type" and "Price" columns contain similar information regarding cost.
#However, the "Price" column, being numeric, is significantly more suitable for quantitative analysis compared to the categorical values
#So Type column is going to be removed
df.drop('Type', axis=1, inplace=True)
```

```python
#View the table
df.head()
df.tail()
```

```python
#Now the data is ready to be exported for analysis
df.to_csv('gplaystore_cleaned.csv', index=False)
```
